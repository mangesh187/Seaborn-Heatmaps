```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
housing = pd.read_csv('Housing.csv')
```

```python
housing.head()
```

```python
housing.info()
```

```python
housing.describe()
```

```python
# simple heatmap 
sns.heatmap(housing.corr())
```

```python
# use of 'annot= True' for the value of corr
sns.heatmap(housing.corr(),
            annot=True
           )
```

```python
#use cmap for changing the color
sns.heatmap(housing.corr(),
            annot=True,
            cmap='viridis'
           )
```

```python
# use of 'fmt='.*g'' it round up the values 

sns.heatmap(housing.corr(),annot=True, 
            cmap='viridis',
            fmt='.1g'
           )
```

```python
# using the parameter 'vmin and vmax'

sns.heatmap(housing.corr(), 
            annot = True,
            vmin=-1, vmax=1,
            fmt='.1g')
```

```python
# using 'center = 0 ' 

sns.heatmap(housing.corr(), 
            annot = True,
            vmin=-1, vmax=1,
            center= 0,
            fmt='.1g'
           )
```

```python
# using 'square = True' 

sns.heatmap(housing.corr(), 
            annot = True,
            square = True,
            vmin=-1, vmax=1,
            center= 0,
            fmt='.1g'
           )
```

```python
# using 'linewidths' and 'cbar_kws' for side bar orientation default is verical 
sns.heatmap(data=housing.corr(),
            annot=True,# default= False 
            linewidths=.5,
            cmap="viridis",
            cbar_kws={"orientation": "vertical"},# default 
            square= False,#default= True
            
           )
```

```python
# using 'linewidths' and 'cbar_kws' for side bar orientation is horizontal 

sns.heatmap(data=housing.corr(),
            fmt='.1g',
            annot=True,# default= False 
            linewidths=.5,
            cmap="viridis",
            cbar_kws={"orientation": "horizontal"},
            square= False,#default= True
            
           )
```

```python
# using 'linecolor'

sns.heatmap(housing.corr(), annot = True,
            vmin=-1, vmax=1,
            fmt='.1g',
            cmap= 'coolwarm',
            linewidths=0.5,
            linecolor='black'
           )
```

```python
# using 'linecolor'

sns.heatmap(housing.corr(),
            annot = True,fmt='.1g',
            vmin=-1, vmax=1, 
            center= 0,
            linewidths=.78, 
            linecolor='white',
            cmap= 'viridis'
           )
```

```python
# using 'mask'= matrix  
# while using matrix 'triu' is used for the lower half of the heatmap

matrix = np.triu(housing.corr())
sns.heatmap(housing.corr(), 
            annot = True,
            mask= matrix,
            fmt='.1g', 
            vmin=-1, vmax=1, 
            center= 0,
            linewidths=.78,
            linecolor='white',
            cmap= 'viridis'
           )
```

```python
#using 'mask'= matrix  
# while using matrix 'tril' is used for the upper half of the heatmap

matrix = np.tril(housing.corr())
sns.heatmap(housing.corr(), 
            annot = True,
            mask= matrix,
            fmt='.1g', 
            vmin=-1, vmax=1, 
            center= 0,
            linewidths=.78,
            linecolor='white',
            cmap= 'viridis'
           )
```
